```python
import pandas as pd
import numpy as np
import os
import tqdm
```

```python
PATH = os.path.expanduser("~/midi/lmd_full")
```

```python
from pathlib import Path
CSV_MIDI_FILES = list(Path(PATH).rglob("*.csv"))
print(f"Number of files found = {len(CSV_MIDI_FILES)}")
```

```python
metadata = {}
count = 0
for csv_file in tqdm.tqdm(CSV_MIDI_FILES):
    try:
        pickle_path = f"{csv_file}.pkl"
        if os.path.exists(pickle_path):
            continue
        
        df = None
        encodings = ["utf-8", "cp1252", "iso-8859-1", "latin1"]
        for encoding in encodings:
            try:
                df = pd.read_csv(csv_file, encoding=encoding, names=list("track time description a b c d".split(' ')), skipinitialspace=True)
                break
            except Exception as _:  # or the error you receive
                pass
        if df is None:
            print(f"Unable to read {csv_file}")
            continue
            
        df = df.applymap(lambda s:s.lower() if type(s) == str else s)
        description_counts = df['description'].value_counts()
        keys = description_counts.keys()
        for key in keys:
            if key in metadata:
                metadata[key] += description_counts[key]
            else:
                metadata[key] = description_counts[key]
        count += 1
        df.to_pickle(pickle_path)
    except Exception as e:
        print(f"Unable to process {csv_file} {e}")
```
